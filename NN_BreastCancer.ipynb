Importing the Dependencies


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows to the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting info about the data
data_frame.info()
```

```python
# Checking for missing values
data_frame.isnull().sum()
```

```python
# Statistical measures about the data
data_frame.describe()
```

```python
# Checking the distribution of Target Values
data_frame['label'].value_counts()
```

1 ----> Benign

0 ---> Malignant

```python
data_frame.groupby('label').mean()
```

Seperating the features and target

```python
x = data_frame.drop(columns = 'label', axis =1)
y = data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

Spliting the data into training & Testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state = 2 )
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
print(x_train_std)
```

Building the Neural Network

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of NN

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation = 'relu'),
                          keras.layers.Dense(2, activation = 'sigmoid')
])
```

```python
# compling the NN
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# Training the NN


history = model.fit(x_train, y_train, validation_split = 0.1, epochs = 30)
```

Visulizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['train', 'validation data'], loc = 'lower right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```
